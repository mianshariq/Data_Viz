```
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Projects/Data/gpa-calculator.csv")
```

```
df=df.head(18)
df
```

```
df=df[["Term", "GPA", "Cum", "nGPA"]]
df.set_index('Term', inplace=True)
df
```

```
df.info()
df
df1=df.head(8)
df1
df2=df.tail(10)
df2
```

```
plt.figure(figsize=(18,6))
sns.lineplot(x=df.index,y=df["GPA"], color="purple", linestyle="--")
sns.lineplot(x=df.index,y=df["Cum"],linestyle="--")
sns.lineplot(x=df2.index,y=df2["nGPA"], color="green", linewidth = 4.5)
ax=plt.gca()
plt.xticks(df.index, rotation=90)
plt.legend(["Term GPA", "Cumulative GPA","Cumulatice GPA Post 2016"], loc=6)
plt.title(" My GPA Story")
ax.set_xlim(0,17)
ax.set_ylim(0,4.2)
ax.grid(False)
ax.axvspan('Spring2010', 'Fall2012', color='red', alpha=0.1, );
ax.axvspan('Fall2016', 'Fall2020', color='green', alpha=0.1, );
plt.annotate('Pre 2016\nCumulative\nGPA 1.96', xy=(7,1.96),xytext=(8,1.0), arrowprops=dict(facecolor="blue",shrink=.01))
plt.annotate('Current Undergraduate\nCumulative\nGPA 2.834', xy=(17,2.8),xytext=(18,2), arrowprops=dict(facecolor="blue",shrink=.01))
plt.annotate('2016-2020\nCumulative\nGPA 3.882 (74 Credits)', xy=(17,3.88),xytext=(18,3), arrowprops=dict(facecolor="blue",shrink=.01))
plt.annotate('Graduated from\nCity College\nGPA 2.716', xy=(14,2.716),xytext=(12,1.5), arrowprops=dict(facecolor="blue",shrink=.01))
plt.annotate('Returned\nto College\nwith 61 Credits', xy=(8,3.85),xytext=(5,3.6), arrowprops=dict(facecolor="blue",shrink=.01))
```
